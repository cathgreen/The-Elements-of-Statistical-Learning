```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(
    "https://web.stanford.edu/~hastie/ElemStatLearn/datasets/prostate.data",
    sep='\t', index_col=0,
)
```

```python
train_idx = df['train'] == 'T'
X = df.drop(columns=['lpsa', 'train']).astype(float)
y = df['lpsa'].astype(float)
```

```python
corr = X[train_idx].corr().round(3)
corr.where(np.tril(np.ones(corr.shape, dtype=bool), k=-1))
```

```python
scaler = StandardScaler()
X = pd.DataFrame(data=scaler.fit_transform(X), index=X.index, columns=X.columns)
mod = sm.OLS(y[train_idx], sm.add_constant(X[train_idx]))
res = mod.fit()
res.summary().tables[1]
```
